Perform the following steps on the titanic dataset:
1. There are missing values in the dataset. Because of that, drop the missing values (Use: dropna). Use columns "Fare" and "Age" to create a scatterplot, then Apply color to the plot based on Gender. Change color of male to blue and change color of female to red. In addtion, chnage marker to triangle (Hint: use seaborn).

```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
import seaborn as sns

titanic = pd.read_csv('train-1.csv')

titanic.dropna()

fare = titanic['Fare']
age = titanic['Age']

x = np.array([age])
y = np.array([fare])

plt.scatter(x,y)
#plt.scatter(x="Age",y="Fare")
plt.show
```

```python
color = ['Red', 'Blue'] # picking the color i want to use
sns.set_palette(color)

sns.palplot(sns.color_palette()) # Prints out the palett in use
```

```python
titanic['Survived'] = titanic['Survived'].map({0:'Female', 1: 'Male'})       # Renaming so that i know wich gender is what
sns.scatterplot(data= titanic,x='Age',y='Fare',hue="Survived",marker='^')
```

2. Use column "Age" to create a histogram.
Change the color of the histogram to green. Add title, xlabel and ylabel to the histogram (Hint: use matplotlib).

```python
plt.hist(titanic['Age'], edgecolor='black', color = 'green')
plt.title('survived')
plt.xlabel('Age')
plt.ylabel('Fare')
```
